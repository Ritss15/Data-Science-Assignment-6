```python
import pandas as pd

df = pd.read_csv("sales_data (5).csv")
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")
```

```python
df.head()
```

```python
df.columns
```

```python
sns.barplot(x="Region", y="Total_Sales", data=df)
plt.title("Total Sales by Region")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots(2, 2, figsize=(12,8))

# 1. Region wise sales
sns.barplot(x="Region", y="Total_Sales", data=df, ax=ax[0,0])
ax[0,0].set_title("Sales by Region")

# 2. Sales distribution
sns.boxplot(x="Region", y="Total_Sales", data=df, ax=ax[0,1])
ax[0,1].set_title("Sales Distribution")

# 3. Profit proxy (Price vs Quantity effect)
sns.scatterplot(x="Quantity", y="Total_Sales", data=df, ax=ax[1,0])
ax[1,0].set_title("Quantity vs Total Sales")

# 4. Sales trend
sns.lineplot(x="Date", y="Total_Sales", data=df, ax=ax[1,1])
ax[1,1].set_title("Sales Trend")

plt.tight_layout()
plt.show()
```

```python
import plotly.express as px

fig = px.bar(
    df,
    x="Region",
    y="Total_Sales",
    color="Region",
    title="Interactive Sales Dashboard"
)

fig.show()
```

 Bar Chart

Shows which region has the highest total sales.

 Box Plot

Displays sales distribution and outliers across regions.

 Heatmap

Shows correlation between Quantity, Price, and Total Sales.

 Line Chart

Represents sales trend over time.

 Interactive Chart

Allows users to explore sales data dynamically.
